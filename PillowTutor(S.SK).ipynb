# 2.3 module : glob ดูชื่อไฟล์ที่ตรงกับนามสกุลไฟล์ที่ระบุ และสร้าง thumbail

```python
from PIL import Image
import glob, os
from termcolor import cprint
size = (128,128)
cprint(glob.glob("*.png"),'yellow',attrs=['bold']) #glob("*.png") จะ return list ที่ของชื่อไฟล์ที่มีนามสกุลตรงกับ format(png)ใน glob ออกมา
for infile in glob.glob("*.png"):
    cprint(infile,'red',attrs=['bold'])
    file, ext = os.path.splitext(infile) #แยก text ออกจากกัน
    cprint(f"file = {file} || ext = {ext}",'blue',attrs=['bold']) #file=ชื่อไฟล์ , ext=นามสกุลของไฟล์   
    im = Image.open(infile)
    im.thumbnail(size)
    im.save(file + ".thumbnail", "JPEG")
```

# 3. Crop รูปภาพ  

```python
from PIL import Image  
  
# เปิดไฟล์รูปภาพในโหมด RGB  
im = Image.open(r"Mulan.png")  
  
# เอาค่า size ความกว้างของรูปภาพเก็บในตัวแปร width เช่นกันความสูงเก็บในตัวแปร height 
width, height = im.size  
  
# เซตบริเวณอาณาเขตที่จะ crop รูปาพ  
left = 6
top = height / 4
right = 174
bottom = 3 * height / 4
  
# Cropped รูปภาพโดยใช้ค่าบริเวณอาณาเขตด้านบน
# รูปภาพ original จะไม่ได้เปลี่ยนไปด้วย
im1 = im.crop((left, top, right, bottom)) 
newsize = (200, 200) 
im1 = im1.transpose(Image.FLIP_LEFT_RIGHT) 
# Shows the image in image viewer(image viewer = โปรแกรมเปิดรูปภาพ)  
im1.show() 
```

# 8. Enhance Image เพิ่มแสง สี 

### from PIL import Image,ImageEngance
### 8.1 ImageEnhance.Color(รูป).enhance(เลขค่า enhance)
### 8.2 ImageEnhance.Contrast(รูป).enhance(เลขค่า enhance)
### 8.3 ImageEnhance.Brightness(รูป).enhance(เลขค่า enhance)
### 8.4 ImageEnhance.Sharpness(รูป).enhance(เลขค่า enhance)
#### Note : ค่า Enhance จะ Default อยู่ที่ 1 ถ้ามากกว่า 1 enhance จะเพิ่ม ถ้าน้อยกว่า 1 enhance จะลด
#### Note : รูป = Image.open("ชื่อรูปภาพ.png/jpg")

```python
from PIL import ImageEnhance,Image
import matplotlib.pyplot as plt 
plt.figure(figsize=(10,10))
im_enhance = Image.open("Mulan.png")
image1 = ImageEnhance.Color(im_enhance).enhance(2.5) #Enhance Color ทำให้สีเข้ม/จาง
plt.subplot(2,2,1)
plt.title("Color Enhance")
plt.imshow(image1)

image2 = ImageEnhance.Brightness(im_enhance).enhance(2.5) #Enhance Brightness เพิ่มความว่าง/ลด
plt.subplot(2,2,2)
plt.title("Color Enhance")
plt.imshow(image2)

image3 = ImageEnhance.Sharpness(im_enhance).enhance(2.5) #Enhance Brightness เพิ่มความคมชัดของภาพ/ลด
plt.subplot(2,2,3)
plt.title("Sharpness Enhance")
plt.imshow(image3)

image4 = ImageEnhance.Contrast(im_enhance).enhance(2.5)#Enhance Contrast เพิ่มการตัดกันของสี/ลด
plt.subplot(2,2,4)
plt.title("Contast Enhance")
plt.imshow(image4)
```
